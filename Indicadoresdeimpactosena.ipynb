```python
# Importar librer√≠as necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Leer el archivo (ya subido)
archivo = 'INDICADORES DE IMPACTO DILIGENCIADO JUAN MA_solo matriz.xlsx'
df = pd.read_excel(archivo)

# Seleccionar y limpiar columnas relevantes
df_limpio = df[['VARIABLE INDICADOR 1', 'VARIABLE CUANTITATIVA E1', 'VARIABLE CUANTITATIVA M1']].copy()
df_limpio.columns = ['Indicador', 'Valor_E1', 'Valor_M1']

# Eliminar filas vac√≠as o con valores faltantes en los campos clave
df_limpio.dropna(subset=['Indicador', 'Valor_E1', 'Valor_M1'], inplace=True)

# Crear gr√°fico de barras comparando valores
plt.figure(figsize=(12, 6))
df_limpio_melted = pd.melt(df_limpio, id_vars='Indicador', var_name='Tipo', value_name='Valor')

# Crear el gr√°fico
sns.barplot(data=df_limpio_melted, x='Indicador', y='Valor', hue='Tipo', palette='Set2')

# Ajustes visuales
plt.title('Comparaci√≥n de Indicadores: Valor E1 vs M1', fontsize=16)
plt.ylabel('Cantidad')
plt.xlabel('Indicador')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()

# Mostrar gr√°fico
plt.show()
```
