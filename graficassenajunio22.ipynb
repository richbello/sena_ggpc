```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Datos resumidos a nivel de ejemplo (sintetizado por grupo). Puedes adaptar desde tu fuente original.
data = {
    'Nivel': ['Alta Dirección'] * 6 + ['Medio'] * 6 + ['Operativo'] * 6,
    'Estrato': [1, 2, 3, 4, 5, 6] * 3,
    'Femenino': [410, 1221, 1738, 842, 352, 164,
                 1169, 3087, 4891, 1749, 509, 107,
                 3373, 3461, 2860, 687, 138, 37],
    'Masculino': [447, 1615, 1868, 783, 332, 141,
                  824, 2388, 3977, 1270, 361, 83,
                  1968, 2495, 1970, 532, 103, 25],
    'Otros': [5, 14, 12, 2, 0, 3,
              18, 22, 37, 12, 1, 0,
              18, 19, 15, 6, 0, 1]
}

df = pd.DataFrame(data)

# Reformatear para seaborn (long-form)
df_long = df.melt(id_vars=['Nivel', 'Estrato'], var_name='Género', value_name='Total')

# Crear gráfico de barras apiladas
plt.figure(figsize=(16, 8))
sns.barplot(data=df_long, x='Estrato', y='Total', hue='Género', hue_order=['Femenino', 'Masculino', 'Otros'],
            estimator=sum, ci=None, palette='Set2', dodge=True)

# Añadir etiquetas de valores encima de cada barra
for index, row in df_long.iterrows():
    if row['Total'] > 0:
        plt.text(x=row['Estrato'] - 1 + (0.25 if row['Género'] == 'Femenino' else (0 if row['Género'] == 'Masculino' else -0.25)),
                 y=row['Total'] + 30,
                 s=int(row['Total']),
                 ha='center', va='bottom', fontsize=8)

plt.title('Participación por Nivel, Estrato y Género', fontsize=16)
plt.xlabel('Estrato')
plt.ylabel('Total de Participantes')
plt.legend(title='Género')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Datos simplificados por Nivel y Estrato
data = {
    'Nivel': ['Alta Dirección'] * 6 + ['Medio'] * 6 + ['Operativo'] * 6,
    'Estrato': [1, 2, 3, 4, 5, 6] * 3,
    'Femenino': [410, 1221, 1738, 842, 352, 164,
                 1169, 3087, 4891, 1749, 509, 107,
                 3373, 3461, 2860, 687, 138, 37],
    'Masculino': [447, 1615, 1868, 783, 332, 141,
                  824, 2388, 3977, 1270, 361, 83,
                  1968, 2495, 1970, 532, 103, 25],
    'Otros': [5, 14, 12, 2, 0, 3,
              18, 22, 37, 12, 1, 0,
              18, 19, 15, 6, 0, 1]
}

# Convertir a DataFrame largo
df = pd.DataFrame(data)
df_long = df.melt(id_vars=['Nivel', 'Estrato'], var_name='Género', value_name='Total')

# Configurar colores elegantes
palette = {
    'Femenino': '#A569BD',  # violeta
    'Masculino': '#3498DB',  # azul
    'Otros': '#F5B041'       # ámbar
}

# Gráfico
plt.figure(figsize=(16, 8))
sns.barplot(
    data=df_long,
    x='Estrato',
    y='Total',
    hue='Género',
    hue_order=['Femenino', 'Masculino', 'Otros'],
    estimator=sum,
    dodge=True,
    errorbar=None,
    palette=palette
)

# Etiquetas encima de cada barra
for index, row in df_long.iterrows():
    if row['Total'] > 0:
        x_pos = row['Estrato'] - 1 + (0.25 if row['Género'] == 'Femenino' else (0 if row['Género'] == 'Masculino' else -0.25))
        plt.text(x=x_pos, y=row['Total'] + 30,
                 s=str(int(row['Total'])), ha='center', va='bottom', fontsize=8)

plt.title('Participación por Estrato, Género y Nivel Ocupacional', fontsize=16)
plt.xlabel('Estrato Socioeconómico')
plt.ylabel('Número de Participantes')
plt.legend(title='Género')
plt.xticks(ticks=range(6), labels=['1', '2', '3', '4', '5', '6'])
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Datos resumidos manualmente de la imagen para ejemplo (se puede cargar de Excel real para análisis completo)
data = {
    "Género": ["Femenino", "Masculino", "Otros"],
    "Alta Dirección": [4727, 5186, 39],
    "Medio": [11513, 8903, 91],
    "Operativo": [10556, 7093, 59],
    "Total": [26796, 21182, 189]
}

df = pd.DataFrame(data)
df.set_index("Género", inplace=True)

# Crear gráfico de barras apiladas
plt.figure(figsize=(10, 6))
colors = ["#008080", "#a39193", "#89ecda"]  # Tonos elegantes de azul

bottom_prev = None
for i, col in enumerate(["Alta Dirección", "Medio", "Operativo"]):
    plt.bar(df.index, df[col], bottom=bottom_prev, label=col, color=colors[i])
    if bottom_prev is None:
        bottom_prev = df[col]
    else:
        bottom_prev += df[col]

# Etiquetas con porcentaje en las barras
totals = df["Total"]
for i, (index, row) in enumerate(df.iterrows()):
    bottom = 0
    for col in ["Alta Dirección", "Medio", "Operativo"]:
        height = row[col]
        percentage = (height / row["Total"]) * 100
        plt.text(i, bottom + height / 2, f"{percentage:.1f}%", ha='center', va='center', color="white", fontsize=9, fontweight="bold")
        bottom += height

# Estética del gráfico
plt.title("Distribución por Género y Nivel Ocupacional", fontsize=14, fontweight='bold')
plt.ylabel("Número de Participantes")
plt.xlabel("Género")
plt.legend(title="Nivel Ocupacional", loc="upper right")
sns.despine()
plt.tight_layout()
plt.show()
```
