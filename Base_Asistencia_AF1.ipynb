```python
from google.colab import files
uploaded = files.upload()
```

```python
import pandas as pd

df = pd.read_excel(list(uploaded.keys())[0])
df.head()
```

```python
df['Dia'] = df['Dia'].astype(str).str.strip()
df['Jornada'] = df['Jornada'].astype(str).str.strip().str.upper()
df['Número de Identificación'] = df['Número de Identificación'].astype(str).str.strip()
```

```python
df_unique = df.drop_duplicates(subset=['Número de Identificación', 'Dia', 'Jornada'])
```

```python
dias_necesarios = {'26', '27', '28'}

asistieron_3_dias = (
    df_unique
    .groupby('Número de Identificación')['Dia']
    .apply(lambda x: dias_necesarios.issubset(set(x)))
)
```

```python
ids_3_dias = asistieron_3_dias[asistieron_3_dias].index.tolist()

df_3dias = df_unique[df_unique['Número de Identificación'].isin(ids_3_dias)]
```

```python
def asistio_doble_jornada(grupo):
    dias = grupo.groupby('Dia')['Jornada'].nunique()
    return all(dias.get(d, 0) == 2 for d in ['26', '27', '28'])

final_ids = []

for id_, grupo in df_3dias.groupby('Número de Identificación'):
    if asistio_doble_jornada(grupo):
        final_ids.append(id_)

df_final = df_3dias[df_3dias['Número de Identificación'].isin(final_ids)]
```

```python
df_final.to_excel("Asistieron_3_dias_2_jornadas.xlsx", index=False)

files.download("Asistieron_3_dias_2_jornadas.xlsx")
```
