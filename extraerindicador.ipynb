```python
from google.colab import files
import pandas as pd

# Subir archivo manualmente
uploaded = files.upload()

for nombre_archivo in uploaded.keys():
    print(f"âœ… Archivo cargado: {nombre_archivo}")
    df = pd.read_excel(nombre_archivo)

    # Normalizar nombres
    df.columns = [str(c).strip().lower() for c in df.columns]

    col_convenio = next((c for c in df.columns if "convenio" in c or "empresa" in c or "conveniente" in c), None)
    col_indicador = next((c for c in df.columns if "indicador" in c), None)
    col_meta_esperada = next((c for c in df.columns if "situaciÃ³n" in c or "esperada" in c), None)
    col_meta_final = next((c for c in df.columns if "avance" in c or "final" in c), None)

    columnas_finales = [col for col in [col_convenio, col_indicador, col_meta_esperada, col_meta_final] if col]
    df_extraido = df[columnas_finales].copy()

    renombres = {
        col_convenio: "Convenio / Empresa",
        col_indicador: "Indicador",
        col_meta_esperada: "Meta (%) SituaciÃ³n Esperada",
        col_meta_final: "Meta (%) Avance FinalizaciÃ³n"
    }
    df_extraido = df_extraido.rename(columns=renombres)

    salida = "Indicadores_Metas.xlsx"
    df_extraido.to_excel(salida, index=False)
    files.download(salida)
    print("ðŸ’¾ Archivo procesado y listo para descargar.")
```
